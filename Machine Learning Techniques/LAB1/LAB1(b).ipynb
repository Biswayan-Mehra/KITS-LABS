```python
import pandas as pd
df = pd.read_csv('iris_EDA.csv')
```

```python
1. Calculate the % of missing values in the columns
```

```python
missing_values=df.isna().sum()
total=len(df)
percentage=(missing_values/total)*10
print(percentage)
```

```python
2. Replace missing value with mean for the numerical column, if the % of missing value is less than 10%. 
```

```python
missing_values=df.isna().sum()
total=len(df)
percentage=(missing_values/total)*10
temp=df.copy()
for i in df.columns:
  if df[i].dtype!="object" and percentage[i]<10:
    c_mean=df[i].mean()
    temp[i].fillna(c_mean,inplace=True)
df.update(temp)
df.head(14)
```

```python
3. Perform the mode imputation for a categorical data
```

```python
missing_values3=df.isna().sum()
total3=len(df)
percentage3=(missing_values3/total3)*10
temp3=df.copy()
for i in df.columns:
  if df[i].dtype =="object" and percentage3[i]<10:
    c_column=df[i].mode().iloc[0]
    temp3[i].fillna(c_column,inplace=True)
df.update(temp3)
df.head(144)
```

```python
4.Perform a KNN Imputer to estimate the missing values
```

```python
from sklearn.impute import KNNImputer
column_names = ['citric acid', 'residual sugar', 'chlorides']
df_missing = df[column_names]
knn_imputer = KNNImputer(n_neighbors=5)
df_imputed_array = knn_imputer.fit_transform(df_missing)
df_imputed = pd.DataFrame(df_imputed_array, columns=column_names)
df[column_names] = df_imputed
print("Missing values imputed using KNN imputer.")
```

```python
5.Drop the columns with more than 10% missing values and display the size 
```

```python
missing_values4=df.isna().sum()
total4=len(df)
percentage4=(missing_values4/total4)*10
temp4=df.copy()
for i in df.columns:
  if percentage4[i]>10:
    temp4.drop(i,axis=1,inplace=True)
print(temp4.head())
print(temp4.shape)
```

```python
 6.Drop the rows with outlier Z-score value > 3 for “Guarantee_Period” and display the  size. 
```

```python
from scipy.stats import zscore
z_scores = zscore(df.select_dtypes(include='number'))
outlier_rows = (z_scores > 3).any(axis=1)
df_no_outliers = df[~outlier_rows]
print(f"Original DataFrame size: {df.shape}")
print(f"DataFrame size after dropping rows with Z-score > 3: {df_no_outliers.shape}")
```

```python
7.Drop the duplicate rows based on more than 50% of column having same value. 
```

```python
columns_to_check =['citric acid', 'residual sugar', 'chlorides']
duplicate_percentage = (df[columns_to_check].duplicated(keep='first').sum() / len(df)) * 100
duplicate_rows = (df[columns_to_check].duplicated(keep='first') / len(columns_to_check)) > 0.5
df_no_duplicates = df[~duplicate_rows]
print(f"Original DataFrame size: {df.shape}")
print(f"DataFrame size after dropping rows with more than 50% duplicate values in specified columns: {df_no_duplicates.shape}")
```

```python
8.Rescale your data using min-max normalization for a numerical feature. (Python 
code) 
```

```python
from sklearn.preprocessing import MinMaxScaler
numerical_feature = 'pH'
numerical_data = df[[numerical_feature]]
scaler = MinMaxScaler()
numerical_data_scaled = scaler.fit_transform(numerical_data)
df[numerical_feature] = numerical_data_scaled
print("DataFrame after min-max normalization:")
print(df)
```

```python
9.Binarize the data by using binarizer class in python
```

```python
from sklearn.preprocessing import Binarizer
numerical_feature = 'pH'
numerical_data = df[[numerical_feature]]
threshold = 9  
binarizer = Binarizer(threshold=threshold)
numerical_data_binarized = binarizer.fit_transform(numerical_data)
df[numerical_feature] = numerical_data_binarized
print("DataFrame after binarization:")
print(df)
```

```python
10.Perform the one-hot encoding for a categorical feature. 
```

```python
categorical_feature = 'total sulfur dioxide'
df_encoded = pd.get_dummies(df, columns=[categorical_feature], prefix=[categorical_feature])
print("DataFrame after one-hot encoding:")
print(df_encoded)
```
