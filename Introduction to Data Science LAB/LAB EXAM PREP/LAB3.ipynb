```python
import os
import pandas as pd
import matplotlib.pyplot as plt
os.chdir('C:/Users/mehra/Codes/KITS/SEM 5/Intro to Data Science/LAB/LAB EXAM PREP/DataSets/')
df = pd.read_csv('Emp_visu.csv')
df.head(2)
```

```python
1. Draw a bar chart with Team and its count (use different colors for each team) 
```

```python
Team = df['Team'].value_counts()
plt.bar(Team.index, Team.values, color=plt.cm.Set3(range(len(Team))))
plt.title('Team-wise Count')
plt.xlabel('Teams')
plt.ylabel('Counts')
plt.xticks(rotation=30, ha='right')
plt.show()
```

```python
2. Draw a comparative bar chart for Salary and New_Salary against each person (first 15 persons)
```

```python
plt.bar(df['First Name'],df['Salary'], width=0.8, label='Salary')
plt.bar(df['First Name'],df['New_Salary'], width=0.4, label='New Salary', align='edge')
plt.xlabel('Persons')
plt.ylabel('Salary')
plt.title('Salary VS New Salary')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
3. Draw a horizontal bar chart for Team and Salary
```

```python
Sal = df.groupby('Team')['Salary'].mean()
plt.barh(Sal.index, Sal.values, color=plt.cm.Set3(range(len(Sal.index))))
plt.xlabel('Salary')
plt.ylabel('Team')
plt.title('Horizontal Bar Chart')
plt.show()
```

```python
4. Draw a stacked bar chart for Salary and New_price against the person (first 10 persons)
```

```python
s = df.head(10)
plt.bar(s['First Name'], s['Salary'], label='Salary')
plt.bar(s['First Name'], s['New_Salary'], label='New Salary', bottom=s['Salary'], color='orange')
plt.xlabel('Person')
plt.ylabel('Salary')
plt.title('Stacked Bar Chart')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
5. Draw a pie chart with Gender and its count
```

```python
gen = df['Gender'].value_counts()
plt.pie(gen.values, labels=gen.index, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.legend()
plt.show()
```

```python
6. Draw the dot plot between person and experience (first 15 persons)
```

```python
s = df.head(15)
plt.plot(s['First Name'], s['Experience'], marker='o', linestyle='')
plt.xlabel('Person')
plt.ylabel('Experience')
plt.title('Dot Plot')
plt.xticks(rotation=90)
plt.show()
```

```python
7. Draw the line plot between age and experience. Observe the trend line.
```

```python
8. Draw the scatter plot between Salary and New_Salary. Observe the correlation
```

```python
plt.scatter(df['Salary'], df['New_Salary'])
plt.xlabel('Salary')
plt.ylabel('New Salary')
plt.title('Scatter Plot')
plt.show()
```

```python
9. Draw the scatter plot between Age and Incentive. Observe the correlation
```

```python
plt.scatter(df['Age'], df['Incentive'])
plt.xlabel('Age')
plt.ylabel('Incentive')
plt.title('Scatter Plot')
plt.show()
```

```python
10. Draw the box plot to show the statistical summary of Age column and verify with describe()
```

```python
df['Age'].fillna(df['Age'].mean())
plt.boxplot(df['Age'])
plt.ylabel('Age')
plt.show()
print(df['Age'].describe())
```

```python
11. Draw the histogram plot for Experience column
```

```python
plt.hist(df['Experience'])
plt.xlabel('Experience')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
plt.hist(df['Experience'], bins=20, density=True)
plt.xlabel('Experience')
plt.ylabel('Probability Density Function')
plt.title('Histogram')
plt.show()
```
