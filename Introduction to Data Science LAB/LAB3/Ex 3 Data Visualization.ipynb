```python
DESCRIPTION: 
    Data visualization uses visual elements like points, lines, bars, colors, and shapes to represent data.
    Common Types of Plots:
        1. Line Plot: A line plot represents data points connected by straight lines, which is useful for visualizing trends or patterns over time or continuous data.
        2. Bar Plot: A bar plot represents categorical data with rectangular bars, where the length of each bar is proportional to the data's value.
        3. Pie Chart: A pie chart represents data as slices of a circular pie, where each slice's size represents the proportion of the data it represents.
        4. Scatter Plot: A scatter plot represents individual data points as dots on a two-dimensional plane, which is useful for identifying relationships between two continuous variables.
```

```python
PROGRAMS:
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("Emp_visu.csv")
```

```python
1. Draw a bar chart with Team and its count
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
teams = data['Team'].value_counts()
plt.bar(teams.index, teams.values, color=plt.cm.Set3(range(len(teams))))
plt.xlabel('Team')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.title('Team-wise Count')
plt.show()
```

```python
2. Draw a comparative bar chart for Salary and New_Salary against each person
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
top_15 = data.head(15)
plt.bar(top_15['First Name'], top_15['Salary'], label='Salary')
plt.bar(top_15['First Name'], top_15['New_Salary'], label='New_Salary')
plt.xlabel('Person')
plt.ylabel('Salary')
plt.title('Salary vs New_Salary')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
3. Draw a horizontal bar chart for Team and Salary
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
team_salary = data.groupby('Team')['Salary'].mean()
plt.barh(team_salary.index, team_salary.values, color='green')
plt.xlabel('Salary')
plt.ylabel('Team')
plt.title('Team-wise Average Salary')
plt.show()
```

```python
4. Draw a stacked bar chart for Salary and New_price against the person
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
top_10 = data.head(10)
plt.bar(top_10['First Name'], top_10['Salary'], label='Salary')
plt.bar(top_10['First Name'], top_10['New_Salary'], bottom=top_10['Salary'], label='New_Salary')
plt.xlabel('Person')
plt.ylabel('Salary')
plt.title('Salary vs New_Salary')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
5. Draw a pie chart with Gender and its count
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
gender_count = data['Gender'].value_counts()
plt.pie(gender_count.values, labels=gender_count.index, autopct='%1.1f%%', colors=['pink', 'lightblue'])
plt.title('Gender Distribution')
plt.show()
```

```python
6. Draw the dot plot between person and experience
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
plt.plot(top_15['First Name'], top_15['Experience'], marker='o', linestyle='')
plt.xlabel('Person')
plt.ylabel('Experience')
plt.title('Dot Plot: Experience of First 15 Persons')
plt.xticks(rotation=45)
plt.show()
```

```python
7. Draw the line plot between age and experience. Observe the trend line.
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
plt.plot(data['Age'], data['Experience'])
plt.xlabel('Age')
plt.ylabel('Experience')
plt.title('Line Plot: Age vs Experience')
plt.show()
```

```python
8. Draw the scatter plot between Salary and New_Salary. Observe the correlation
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
plt.scatter(data['Salary'], data['New_Salary'])
plt.xlabel('Salary')
plt.ylabel('New_Salary')
plt.title('Scatter Plot: Salary vs New_Salary')
plt.show()
```

```python
9. Draw the scatter plot between Age and Incentive. Observe the correlation
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
plt.scatter(data['Age'], data['Incentive'])
plt.xlabel('Age')
plt.ylabel('Incentive')
plt.title('Scatter Plot: Age vs Incentive')
plt.show()
```

```python
10. Draw the box plot to show the statistical summary of Age column and verify with
describe()
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
plt.boxplot(data['Age'])
plt.ylabel('Age')
plt.title('Box Plot: Age')
plt.show()
```

```python
11. Draw the histogram plot for Experience column
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
plt.hist(data['Experience'])
plt.xlabel('Experience')
plt.ylabel('Frequency')
plt.title('Histogram: Experience')
plt.show()
```

```python
12. Draw the histogram plot for Experience column with bin value and PDF
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
plt.hist(data['Experience'], bins=10, density=True)
plt.xlabel('Experience')
plt.ylabel('Probability Density')
plt.title('Histogram: Experience (PDF)')
plt.show()
```
