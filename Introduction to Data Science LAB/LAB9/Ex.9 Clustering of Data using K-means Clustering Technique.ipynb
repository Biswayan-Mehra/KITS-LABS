```python
AIM: To demonstrate Clustering of Data using K-means Clustering Technique.
```

```python
DESCRIPTION:
    1. K-Means Clustering:
        K-means clustering is a type of unsupervised machine learning algorithm used for partitioning data into K distinct, non-overlapping clusters based on similarity or proximity.
    2. Elbow Method:
        The Elbow Method is a heuristic used to determine the optimal number of clusters (K) in a dataset. It involves plotting the within-cluster sum of squares (WCSS) for different values of K and selecting the "elbow" point on the plot, which represents a balance between minimizing WCSS and avoiding overfitting.
    3. Within-Cluster Sum of Squares (WCSS):
        WCSS is a measure used to evaluate the quality of clustering in K-means. It calculates the sum of squared distances between data points and their assigned cluster centroids. Lower WCSS indicates tighter clusters.
    4. Cluster Centroids:
        Cluster centroids are the central data points in each cluster. They represent the average location of data points within the cluster and are used to characterize and visualize the clusters.
    5. Silhouette Score:
        The Silhouette Score is a measure of how similar an object is to its own cluster compared to other clusters. It ranges from -1 to 1, where a high score indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters.
    6. Davies-Bouldin Score:
        The Davies-Bouldin Score is a measure used to evaluate the average similarity between each cluster and its most similar cluster. A lower Davies-Bouldin Score indicates better clustering; a score of 0 suggests perfect clustering.
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
data = pd.read_csv('mall_customers.csv')
data.head(3)
```

```python
a. Use the columns: ' Age', 'Annual_Income'as the input variables.
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")

X = data[['Age', 'Annual_Income']]
```

```python
b. Compute the optimal number of clusters ‘K’ from 1-10 using Elbowmethod.
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,max_iter=300, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
c. Plot the graph between number of cluster K and within-cluster sumof squares (WCSS)
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS')
plt.show()
```

```python
d. value.
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k,max_iter=300, n_init=10)
kmeans.fit(X)
```

```python
e. Perform the K-means clustering with optimal K. 
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
cluster_centers = kmeans.cluster_centers_
print("Cluster Centroids:")
print(cluster_centers)
```

```python
f. Display the cluster centroids.
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
plt.figure(figsize=(8, 6))
plt.scatter(X['Age'], X['Annual_Income'], c=kmeans.labels_, cmap='rainbow')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=100, c='black', label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.legend()
plt.show()
```

```python
g. Visualize the data representation of K-means clustering. 
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
plt.figure(figsize=(8, 6))
plt.scatter(X['Age'], X['Annual_Income'], c=kmeans.labels_, cmap='rainbow')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=100, c='black', label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.legend()
plt.show()
```

```python
h. Change the value of K in K-means with different values and tabulate the silhouette_score and davies_bouldin_score obtained. 
```

```python
print("BISWAYAN MEHRA \nURK21CS1004")
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, n_init=10)
    kmeans.fit(X)
    silhouette = silhouette_score(X, kmeans.labels_)
    davies_bouldin = davies_bouldin_score(X, kmeans.labels_)
    print(f'K={k}: Silhouette Score={silhouette:.4f}, Davies-Bouldin Score={davies_bouldin:.4f}')
```
